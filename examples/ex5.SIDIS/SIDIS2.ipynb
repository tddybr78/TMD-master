##compute bT*W for DY1 and store values in BTW1

```python
def PDFNP(x,zetaF,bT2):
    arg = -0.25*(0.19*np.log(zetaF/(1.0)))*bT2
    arg+= -1.0*((0.013*(1-x)**3)/x)*bT2
    return np.exp(arg)
```

```python
def FFNP(z,zetaD,bT2):
    arg = -0.25*(0.19*np.log(zetaD/(1.0)))*bT2
    arg+= -(0.1)*bT2                                           
    return np.exp(arg)
```

```python
sidis.TPDF.D['NP model'] = PDFNP
sidis.TFF.D['NP model'] = FFNP
```
