#Testing the CTEQ6 wrapper

```python
# libraries
import sys,os
sys.path.insert(1,'../../')
import pylab as py
import numpy as np
%matplotlib inline  
from tools import ProgressBar,tex
from cteq6 import cteq6
```

```python
# load cteq6l table
cteq6.setctq6(1,'../../cteq6/')

# eval u and d
Q=2
X=10**np.linspace(-4,0)
XU=[x*cteq6.ctq6pdf(1,x,Q) for x in X]
XD=[x*cteq6.ctq6pdf(2,x,Q) for x in X]

# make plot
ax=py.subplot(111)
ax.plot(X,XU,label=tex('xu'))
ax.plot(X,XD,label=tex('xd'))
ax.set_ylim(1e-3,1e1)
ax.semilogx()
ax.semilogy()
ax.set_ylabel(tex('xf(x)'),size=20)
ax.set_xlabel(tex('x'),size=20)
ax.set_title(tex('Q=%0.2f GeV'%Q))
ax.legend()
```
