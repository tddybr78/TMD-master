# TMD PDF

```python
# libraries
import sys,os
sys.path.insert(1,'../../')
import pylab as py
import numpy as np
%matplotlib inline  
from TMDPDF import TMDPDF
from StrongCoupling import StrongCoupling
from tools import ProgressBar,tex
```

# Alpha Strong

```python
# testing alphaS
SC=StrongCoupling()
MU2=np.linspace(5,1000)
alphaS=[SC.get_alphaS(mu2) for mu2 in MU2]
ax=py.subplot(111)
ax.plot(MU2,alphaS)
ax.set_xlabel(tex('Q2'),size=20)
ax.set_ylabel(tex(r'\alpha_s'),size=20)
```

# PDF in bT space

```python
# begin TMD PDF class
tmd=TMDPDF()

# set params
x=0.5
mu2=5
xiF=mu2

# compute
BT=10**np.linspace(-3,1)
PDF=[bT*tmd.get_PDF_bT_space(x,bT**2,mu2,xiF) for bT in BT]

# make plot
ax=py.subplot(111)
ax.plot(BT,PDF,'r-')
ax.set_ylabel(tex('bT*PDF(bT)'),size=20)
ax.set_xlabel(tex('bT'),size=20)
```

# PDF in qT space

```python
# begin TMD PDF class
tmd=TMDPDF()

# set params
x=0.5
mu2=5
xiF=mu2

# compute 
QT=np.linspace(0.5,3)
PDF=[tmd.get_PDF(x,mu2,xiF,qT) for qT in QT]

# make plot
ax=py.subplot(111)
ax.plot(QT,PDF,'r-')
ax.set_ylabel(tex('PDF(qT)'),size=20)
ax.set_xlabel(tex('qT'),size=20)
```
