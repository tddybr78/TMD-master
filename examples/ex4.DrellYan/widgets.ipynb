```python
def plot(**kwargs):
    ax=py.subplot(111)
    X=np.linspace(0,np.pi)
    p0=kwargs['p0']
    p1=kwargs['p1']
    p2=kwargs['p2']
    global p0
    global p1
    global p2
    exec kwargs['f']
    ax.plot(X,func(X),'r-')
    ax.set_ylim(-6,6)
interact(plot,p0=(0.5,10,0.01),p1=(0,10,0.01),p2=(-0.5,0,0.01),
         f='func=lambda x: p0*np.sin(p1*x)*np.exp(p2*x)')
```
