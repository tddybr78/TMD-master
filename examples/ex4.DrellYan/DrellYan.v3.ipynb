# TMD PDF

```python
# libraries
import sys,os
sys.path.insert(1,'../../')
import pylab as py
import numpy as np
%matplotlib inline  
from TMDPDF import TMDPDF
from StrongCoupling import StrongCoupling
from tools import ProgressBar,tex
from DrellYan import DrellYan
```

```python
# initialize DY 
DY=DrellYan(pdfset='fake')
#DY=DrellYan(pdfset='cteq6')

# setup params
Q=91.1876
RS=1900.0
y=0 

# define bT values to be computed
BT=np.linspace(1e-3,1.4,100)

# compute bT*W for set I 
DY.TMD.set_NP_params(Set=1)
bar=ProgressBar(BT.size,'computing DY set 1')
BTW_set1=[]
cnt=0
for bT in BT: 
    BTW_set1.append(bT*DY.get_L_bT_space(Q,RS,y,bT))
    bar.animate(cnt)
    cnt+=1

# compute bT*W for set I 
DY.TMD.set_NP_params(Set=2)
bar=ProgressBar(BT.size,'computing DY set 2')
BTW_set2=[]
cnt=0
for bT in BT: 
    BTW_set2.append(bT*DY.get_L_bT_space(Q,RS,y,bT))
    bar.animate(cnt)
    cnt+=1

# make plot
py.figure(figsize=(8.5,5.5))
ax=py.subplot(111)

ax.plot(BT,BTW_set1,'m-.',lw=2,label=tex('b_{max}=%0.2f')%(DY.TMD.D['bT2max']**0.5))
ax.plot(BT,BTW_set2,'b--',lw=2,label=tex('b_{max}=%0.2f')%(DY.TMD.D['bT2max']**0.5))

ax.set_ylabel(tex('b_T\\tilde{W}\;(nb/GeV)'),size=20)
ax.set_xlabel(tex('b_T(GeV^{-1})'),size=20)
ax.legend()
```
